```python
# Normalize tables

def normalize_table(title, movie_id, df, column):
    result = df[column].apply(pd.read_json)
    titles = df[title]
    ids = df[movie_id]
    for i in range(len(df)):
        result[i][title] = titles[i]
        result[i][movie_id] = ids[i]
    return pd.concat(result.values).reset_index().drop("index", axis = 1)
```

```python
# 1. For each movie, compute the number of cast members

cast = normalize_table("title", "movie_id", cr, "cast")
cast.rename(columns={'name':'actor_name'}, inplace=True)
actors_each_movie = cast.groupby(by = ['movie_id','title'], as_index = False).size().to_frame().reset_index().rename(columns = {0:'Number of actors'})
actors_each_movie
```

```python
# 2. How many movies do not have a homepage?

mv1 = mv.dropna(subset = ["homepage"])
movies_no_homepage = len(mv) - len(mv1) 
movies_no_homepage
```

```python
# 3. For each year, how many movies do not have a homepage?

mv2 = mv[mv.homepage.isnull()]
mv2["year"] = mv2["release_date"].str.extract("(.*)-.*-.*", expand = True)
mv2.groupby("year").size()
```

```python
# 4. Extract the domain of each homepage

def extract_domain(hp):
    domains = []
    homepage = hp.split(" ")
    for i in range(len(homepage)):
        domains.append(homepage[i].split('/')[2].split('.')[-1])
    return str(domains)[1:-1].replace("\'", "")
   
mv1["domains"] = mv1["homepage"].apply(extract_domain)
mv1[["original_title", "homepage", "domains"]]
```

```python
# (esempio) interrogare il dataset: qual è il domain per il film Cargo?

mv1[["original_title", "homepage", "domains"]][mv1["original_title"] ==  "Cargo"]

# CASO PARTICOLARE = il film Cargo presenta due domain!!!
```

```python
# 5. Extract a set of normalized tables. (la tabella cast è stata già creata per risolvere il punto 1)

crew = normalize_table("title", "movie_id", cr, "crew")
crew.rename(columns={'name':'crew_member_name'}, inplace=True)

genres = normalize_table("original_title", "id", mv, "genres")
genres.rename(columns={'name':'film_genres'}, inplace=True)

keywords = normalize_table("original_title", "id", mv, "keywords")
keywords.rename(columns={'name':'film_keyword'}, inplace=True)

production_companies = normalize_table("original_title", "id", mv, "production_companies")
production_companies.rename(columns={'name':'production_companies'}, inplace=True)

production_countries = normalize_table("original_title", "id", mv, "production_countries")
production_countries.rename(columns={'iso_3166_1':'country_id', 'name':'country'}, inplace=True)

spoken_languages = normalize_table("original_title", "id", mv, "spoken_languages")
spoken_languages.rename(columns={'iso_639_1':'language_id', 'name':'language'}, inplace=True)

# Create normalized copies of the original dataframes
mv_norm = mv.drop(['genres', 'keywords', 'production_companies', 'production_countries', 'spoken_languages'],axis = 1)
cr_norm = cr.drop(['cast', 'crew'], axis = 1)
```

```python
#10. For each company and each decade, compute the overall revenue

mv_norm["year"] = mv_norm["release_date"].str.extract("(.*)-.*-.*", expand = True)

def estract_decade(year):
    return str(year)[:3] + '0'

mv_norm['decade'] = mv_norm['year'].apply(estract_decade)

mv_norm_companies = production_companies.merge(right = mv_norm[['id', 'revenue','decade']], on = 'id')

mv_norm_companies.groupby(by = ['production_companies','decade'], as_index = False).sum().drop('id', axis = 1)
```

```python
# (esempio) interrogare il dataset: qual è il ricavo complessivo della compagnia 21 Laps Entertainment per decade di attività? 

mv_norm_companies[mv_norm_companies['production_companies']=='21 Laps Entertainment'].groupby(by=['production_companies','decade']).sum().drop('id', axis = 1)
```

```python
# (esempio) interrogare il dataset: qual è il ricavo complessivo della compagnia Walt Disney Pictures per decade di attività? 

mv_norm_companies[mv_norm_companies['production_companies']=='Walt Disney Pictures'].groupby(by=['production_companies','decade']).sum().drop('id', axis = 1)
```
