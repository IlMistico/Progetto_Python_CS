```python
import pandas as pd
import numpy as np
mv = pd.read_csv('C:/Users/Paolo/Desktop/Computer Science (Python)/Progetto/tmdb_5000_movies.csv')
cr = pd.read_csv('C:/Users/Paolo/Desktop/Computer Science (Python)/Progetto/tmdb_5000_credits.csv')
```

```python
#Questa funzione costruisce i dataframe a partire dalle stringhe formato json memorizzate in alcune colonne.
#Lo scopo è normalizzare le tabelle

def normalize_table(title, movie_id, df, column):
    result = df[column].apply(pd.read_json)
    titles = df[title]
    ids = df[movie_id]
    for i in range(len(df)):
        result[i][title] = titles[i]
        result[i][movie_id] = ids[i]
    return pd.concat(result.values).reset_index().drop("index", axis = 1)
```

```python
#1.For each movie, compute the number of cast members
cast = normalize_table("title", "movie_id", cr, "cast")
actors_each_movie = cast.groupby(by = ['movie_id','title'], as_index = False).size()
actors_each_movie
```

```python
# 2. How many movies do not have a homepage?
len(mv[mv['homepage'].isnull()].dropna(subset = ['release_date']))
```

```python
#3. For each year, how many movies do not have a homepage?
def extract_year(timestamp):
    if timestamp is np.nan:
        return np.nan
    else:
        return timestamp.year

mv2 = mv.dropna(subset = ['release_date'])
mv2['release_year'] = mv2['release_date'].apply(pd.to_datetime).apply(extract_year) #Questo assegnamento crea un warning,perchè?
movies_per_year = mv2[mv2['homepage'].isnull()].groupby(by='release_year').size()
movies_per_year
```

```python
#4. Extract the domain of each homepage (for example .com, .it, .org)

def extract_domain(hp):
    domains = []
    if hp is np.nan:
        return np.nan
    else:
        homepage = hp.split(' ')
        for i in range(len(homepage)):
            domains.append(homepage[i].split('/')[2].split('.')[-1])
        return str(domains)[1:-1].replace("\'", "")
    
domains = mv['homepage'].apply(extract_domain).to_frame()
domains_with_titles = mv['title'].to_frame().join(domains).rename(columns = {'homepage':'domain'})
domains_with_titles
```

```python
#5. Extract a set of normalized tables. 
#That is, each entry of a normalized table must contain exactly one value (not a list or a dictionary).

# create table(cast) from cr
#già creata preliminarmente per risolvere il punto 1

# create table(crew) from cr
crew = normalize_table("title", "movie_id", cr, "crew")

# create table(genders) from mv
genders = normalize_table("original_title", "id", mv, "genres")

# create table(keyword) from mv
keywords = normalize_table("original_title", "id", mv, "keywords")

# create table(production_companies) from mv
production_companies = normalize_table("original_title", "id", mv, "production_companies")

# create table(production_countries) from mv
production_countries = normalize_table("original_title", "id", mv, "production_countries")

# create table(spoken_languages) from mv
spoken_languages = normalize_table("original_title", "id", mv, "spoken_languages")

#create normalized copies of the original dataframes
mv_norm = mv.drop(['genres', 'keywords', 'production_companies', 'production_countries', 'spoken_languages'],axis = 1)
cr_norm = cr.drop(['cast', 'crew'], axis = 1)
```

```python
#6. For each movie, compute the gross margin (difference between revenue and budget)
mv_norm['gross_margin'] =  mv_norm['revenue'] - mv_norm['budget']
mv_norm[['title','gross_margin']]
```

```python
#7. For each movie, compute the number of crew members
crew.groupby(by = 'title').size()
```

```python
#8. For each movie, compute the number of directors
crew[crew['job']=='Director'].groupby(by = 'title').size().to_frame()
```

```python
#9. For each language, compute the number of movies where such language is spoken.
spoken_languages.groupby(by = 'name').size().to_frame().reset_index().rename(columns = {'name': 'language', 0:'Number of movies where such language is spoken'})
```

```python
#10. For each company and each decade, compute the overall revenue
def estract_decade(year):
    return str(year)[:3] + '0'

mv_norm['release_decade'] = mv_norm['release_date'].apply(pd.to_datetime).apply(extract_year).apply(estract_decade)

companies_upgraded = production_companies.merge(right = mv_norm[['id', 'revenue','release_decade']], on = 'id')
companies_upgraded.groupby(by = ['name', 'release_decade']).sum().drop('id', axis = 1)
```

```python
#11. For each decade, compute the company with maximum revenue
companies_upgraded.iloc[companies_upgraded[['name','release_decade','revenue']].groupby(by ='release_decade', as_index = False).idxmax()['revenue']]
```

```python
#12. In each year, how many movies have revenue smaller than the budget?
mv_norm['release_year'] = mv_norm['release_date'].apply(pd.to_datetime).apply(extract_year)
mv_norm[mv_norm['gross_margin'] < 0].groupby(by = 'release_year').size()
```
