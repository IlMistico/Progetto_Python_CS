```python
# può movie_id essere utilizzato come indice univoco?

def check_uniqueness(df, column):
    return len(df[column]) == len(df[column].drop_duplicates())
check_uniqueness(cr,'movie_id') and check_uniqueness(mv,'movie_id')
```

```python
# Funzione per estrarre i df scritti in json
# result è una Series di Dataframe
# movie_id è una Series degli ID dei film, nello stesso ordine dei dataframe contenuti in result
# si attacca ad ogni dataframe result l'id del film a cui si riferisce
# dalla Series di Dataframe ricavo un unico Dataframe

def extract_json(df, column):
    result = df[column].apply(pd.read_json) 
    movie_id = df['movie_id'] 
    for i in range(len(df)):
        result[i]['movie_id'] = movie_id[i] 
    return pd.concat(result.values).reset_index().drop("index", axis = 1) 
```

```python
# Si estraggono tutti i json e vengono eliminate le corrispondenti colonne nei df iniziali

cast = extract_json(cr,'cast')

crew = extract_json(cr,'crew')

genres = extract_json(mv,'genres')

keywords = extract_json(mv,'keywords')

production_companies = extract_json(mv,'production_companies')

production_countries = extract_json(mv,'production_countries')

spoken_languages = extract_json(mv,'spoken_languages')


movies = mv.drop(['genres', 'keywords', 'production_companies', 'production_countries', 'spoken_languages'],axis = 1)

credits = cr.drop(['cast', 'crew'], axis = 1)
```

```python
# 4. Extract the domain of each homepage

def extract_domain(hp):
    domains = []
    homepage = hp.split(" ")
    for i in range(len(homepage)):
        domains.append(homepage[i].split('/')[2].split('.')[-1])
    return str(domains)[1:-1].replace("\'", "")
   
mv1["domains"] = mv1["homepage"].apply(extract_domain)
mv1[["original_title", "homepage", "domains"]]
```

```python
# 10. For each company and each decade, compute the overall revenue

movies["year"] = movies["release_date"].str.extract("(.*)-.*-.*", expand = True)

def estract_decade(year):
    return str(year)[:3] + '0'

movies['decade'] = movies['year'].apply(estract_decade)
movies_companies_id = production_companies.merge(right = movies[['movie_id','revenue','decade']], on = 'movie_id')
movies_companies_name = production_companies_name.merge(right = movies_companies_id[['company_id','revenue','decade']], on = 'company_id')
overall_companies_revenue = movies_companies_name.groupby(by = ['company_id','name','decade'], as_index = False).sum()
overall_companies_revenue
```

```python
# 13. Distribute the revenue according to the order of appearance in a movie. 
# Assume that the i-th actor contributes twice as much as the (i+1)-th actor to the revenue.

movies_revenue = movies_companies_id[['movie_id','revenue']].drop_duplicates()
cast2 = pd.DataFrame()
cast2 = cast.merge(right = actors_each_movie_id.reset_index(), on = 'movie_id').merge(right = movies_revenue, on = 'movie_id')
cast2 = cast2[['actor_id','movie_id','order','number of actors','revenue']].rename(columns = {'number of actors':'num_actors'})


# legge di distribuzione: 2^(n-k-1)/(2^n -1) * revenue = [1/2^k * (1 + 1/(2^n -1))] * revenue = revenue/2^k + revenue/(2^k * (2^n -1))
# la funzione utilizza direttamente le Series con i valori necessari

def assign_part(orders,num_actors,revenues):
   
    part = [0]*len(orders)
    for i in range(len(orders)):
        k = orders[i]
        n = num_actors[i]
        rev = revenues[i]
        e_n = np.exp2(n)
        e_k = np.exp2(k)
        
        number = rev/e_k + rev/(e_k * (e_n -1))
        part[i] = np.round(number/2)
        
    return part
cast2['part'] = assign_part(cast2['order'],cast2['num_actors'],cast2['revenue'])

pd.set_option('display.float_format', lambda x: '%.2f' % x)
cast2
```

```python
# Check: è vero che la somma delle parti dei vari attori corrisponde alla revenue totale?

movies_id = cast2['movie_id'].unique()
for movie_id in movies_id:
    print(cast2[cast2['movie_id'] == movie_id]['revenue'].iloc[0], cast2[cast2['movie_id'] == movie_id]['part'].sum())
```
